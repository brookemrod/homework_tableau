# Citi Bike

Import Data

```python
import pandas as pd
```

```python
df201806 = pd.read_csv("JC-201806-citibike-tripdata.csv")
df201807 = pd.read_csv("JC-201807-citibike-tripdata.csv")
df201808 = pd.read_csv("JC-201808-citibike-tripdata.csv")
df201809 = pd.read_csv("JC-201809-citibike-tripdata.csv")
df201810 = pd.read_csv("JC-201810-citibike-tripdata.csv")
df201811 = pd.read_csv("JC-201811-citibike-tripdata.csv")
df201812 = pd.read_csv("JC-201812-citibike-tripdata.csv")
df201901 = pd.read_csv("JC-201901-citibike-tripdata.csv")
df201902 = pd.read_csv("JC-201902-citibike-tripdata.csv")
df201903 = pd.read_csv("JC-201903-citibike-tripdata.csv")
df201904 = pd.read_csv("JC-201904-citibike-tripdata.csv")
df201905 = pd.read_csv("JC-201905-citibike-tripdata.csv")
df201806.head()
```

```python
df201905.head()
```

```python
# Concatenate JC dataframes
df = pd.concat([df201806, df201807, df201808, df201809, df201810, df201811, df201812, \
           df201901, df201902, df201903, df201904, df201905])
df.head()
```

Explore Data

```python
df.describe()
```

```python
pctg = round((len(df201806) / len(df201905) - 1)*100, 1)
print(f"From 2018-06 to 2019-05, total ridership has grown by {pctg}%.")
```

```python
sub201806 = df201806["usertype"].value_counts()["Subscriber"]
print("2018-06")
print(f"Proportion of subscribers: {sub201806 / df201806['usertype'].count()}")
cus201806 = df201806["usertype"].value_counts()["Customer"]
print(f"Proportion of customers: {cus201806 / df201806['usertype'].count()}")

sub201905 = df201905["usertype"].value_counts()["Subscriber"]
print("---\n2019-05")
print(f"Proportion of subscribers: {sub201905 / df201905['usertype'].count()}")
cus201905 = df201905["usertype"].value_counts()["Customer"]
print(f"Proportion of customers: {cus201905 / df201905['usertype'].count()}")
```

```python
summer = pd.concat([df201806, df201807, df201808])
summer["hour"] = pd.to_datetime(summer.starttime)
summer["hour"] = summer["hour"].dt.hour
summer.groupby(["hour"])["bikeid"].count()
```

```python
winter = pd.concat([df201812, df201901, df201902])
winter["hour"] = pd.to_datetime(winter.starttime)
winter["hour"] = winter["hour"].dt.hour
winter.groupby(["hour"])["bikeid"].count()
```

```python
# Query date from the dataframe
d='2019-05-01' 
today_date=pd.to_datetime(d).date()

df201905["date"] = pd.to_datetime(df201905["starttime"])
df201905["date"] = df201905["date"].dt.date
df_queried_date = df201905.loc[df201905["date"] == today_date]
```

```python
# Find top stations
df_top_stations = df_queried_date.groupby(["start station id"])\
    ["start station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_stations = df_top_stations.rename(columns={"start station id":"number of start stations"})
df_top_stations["start station id"] = df_top_stations.index
df_top_stations.reset_index(drop=True, inplace=True)
# df_top_stations
```

```python
# Report top station names
stations = df201809[["start station id", "start station name"]].drop_duplicates()
df_top_stations.merge(stations, how="inner", on="start station id")
```

```python
# Find top stations
df_top_stations = df_queried_date.groupby(["end station id"])\
    ["end station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_stations = df_top_stations.rename(columns={"end station id":"number of end stations"})
df_top_stations["end station id"] = df_top_stations.index
df_top_stations.reset_index(drop=True, inplace=True)
# df_top_stations
```

```python
# Report top station names
stations = df201809[["end station id", "end station name"]].drop_duplicates()
df_top_stations.merge(stations, how="inner", on="end station id")
```
